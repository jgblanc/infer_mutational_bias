```python
filepath = "../data/small.tsv"
da_dict = {}
daf_dict = {}
with open(filepath) as fp:
    line = fp.readline()
    while line:
        line = line.strip()
        line_list = line.split()
        if line_list[11] == "derived_ancestral":
            key = line_list[0] + ":" + line_list[1] 
            key_value_allele = line_list[9]
            key_value_daf = line_list[19]
            da_dict[key] = key_value_allele
            daf_dict[key] = key_value_daf
        line = fp.readline()
```

```python
filepath_GWAS = "../output/GWAS_ATLAS/parsed_gwas/small.txt"
new_file_lines = []
with open(filepath_GWAS) as fp: 
    line = fp.readline()
    while line:
        line = line.strip()
        line_list = line.split()
        SNP = line_list[0]
        print(SNP)
        EA = line_list[1]
        RA = line_list[11]
        EAF = line_list[8]
        MAF = line_list[9]
        RAF = line_list[10]
        try: 
            DA = da_dict[SNP]
            print(DA)
            print(EA)
            DAF = float(daf_dict[SNP])
            if MAF != "NA" and EAF == "NA":
                if DAF < 0.5:
                    if DA == EA:
                        line_list[8] = MAF
                    else:
                        line_list[8] = str(1 - float(MAF)) 
                if DAF >= 0.5:
                    if DA == EA:
                        line_list[9]= str(1 - float(MAF)) 
                    else:
                        line_list[8] = MAF
                if RA == EA:
                    line_list[10] = line_list[8]
                else:
                    line_list[10] = str(1 - float(line_list[8]))
    
            if DA == EA: 
                new_entry = "T"
                if EAF == "NA":
                    line_list[8] = DAF 
                daf = str(line_list[8])
            else: 
                new_entry = "F" 
                if EAF == "NA":
                    line_list[8] = str(1 - DAF)
                daf = str(1 - float(line_list[8]))
            if DA == RA:
                r_derived = "T"
                if RAF == "NA":
                    line_list[10] = str(DAF)
            else:
                r_derived = "F"
                if RAF == "NA":
                    line_list[10] = str(1 - DAF)
            if MAF == "NA":
                if DAF < 0.5:
                    line_list[9] = str(DAF)
                else:
                    line_list[9] = str(1-DAF)
        except KeyError:
            new_entry = "NA" 
            daf = str("NA")
            r_derived = "NA"
        print(daf)
        line_list.append(new_entry)
        line_list.append(daf)
        print(line_list)
        line_list.append(r_derived)
        new_line = ",".join(line_list)
        new_file_lines.append(new_line)
        line = fp.readline()
        
```
